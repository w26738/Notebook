** Matplotlib 101 **

### Basics

#### import matplotlib and see the version

```python
import matplotlib

print(matplotlib.__version__)
```

#### import pyplot module and draw a curve

```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline

x = np.arange(100)
y = x * x

plt.plot(x, y)
```

#### plot multiple curves in one figure

```python
x = np.linspace(-np.pi, np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1)
plt.plot(x, y2)
```

#### scatter plot

```python
data = np.random.rand(2, 512)
print(data)
plt.scatter(data[0], data[1])
```

#### plot multiple bar charts

```python
data = np.random.randint(0, 100, size=(3, 4))
x = np.arange(4)

plt.bar(x, data[0], color='r', width=0.25)
plt.bar(x+0.25, data[1], color='g', width=0.25)
plt.bar(x+0.5, data[2], color='b', width=0.25)
```

#### plot stack bar charts

```python
data = np.random.randint(0, 100, size=(3, 4))
colors = ['r', 'g', 'b']

rows = data.shape[0]
columns = np.arange(data.shape[1])
for row in range(rows):
    plt.bar(columns, data[row], bottom=np.sum(data[:row], axis=0), color=colors[row%len(colors)])
```

#### plot error bar

```python
size = 10
x = np.linspace(-np.pi, np.pi, size)
y = np.sin(x) + np.random.random(size)

plt.errorbar(x, y, yerr=1, fmt='.r', capsize=5, capthick=3)
```

#### plot pie charts

```python
labels = ['priority 1', 'priority 2', 'priority 3', 'priority 4']
colors = ['#996600', '#CCCC66', '#666600', '#CC3366']
fractions = [45, 30, 15, 10]
explode =(0, 0, 0, 0.05)
distance = 1.1

plt.pie(fractions, explode=explode, colors=colors, labeldistance=distance, labels=labels)
plt.axis('equal')
```

#### plot histograms

```python
np.random.seed(100)
data = np.random.randn(1000)
kwargs = dict(alpha=0.5, color='blue', bins=20, edgecolor='black')

plt.hist(data, **kwargs)
```

#### plot Kernel-Density-Estimation (KDE)

```python
import seaborn as sns

np.random.seed(100)
data = np.random.randn(1000)

sns.kdeplot(data)
```

#### plot boxplots

```python
np.random.seed(100)
data = np.random.randn(1000, 4)

plt.boxplot(data, sym='');
```

### Color and Style

#### creating a random scatter plot with points of many colors and sizes

```python
data = np.random.random(size=(2,100))
colors = np.random.random(100)
size = 1000 * np.random.random(100)

# return a list of all registered colormaps
plt.colormaps()
plt.scatter(data[0], data[1], s=size, c=colors, alpha=0.25, cmap='winter')
plt.colorbar()
```
